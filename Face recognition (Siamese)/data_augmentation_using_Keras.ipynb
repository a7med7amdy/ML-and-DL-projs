```python
# Importing necessary functions 
from keras.preprocessing.image import ImageDataGenerator,  array_to_img, img_to_array, load_img 
import os
```

```python
datagen = ImageDataGenerator( 
#         rotation_range = 40, 
        shear_range = 0.2, 
        zoom_range = 0.2, 
        horizontal_flip = True, 
        brightness_range = (0.5, 1.5))

path= "/home/ahmed/intern work/actor web scrapping/augmentation/actors faces dataset/men"

actors= os.listdir(path)

for actor in actors:
    print(actor)
    imgs = os.listdir(path+"/"+actor)
    for image in imgs:
        img = load_img(path+"/"+actor+"/"+image)  
        x = img_to_array(img) 
        x = x.reshape((1, ) + x.shape)  
        i = 0
        for batch in datagen.flow(x, batch_size = 1, 
                                  save_to_dir = path+"/"+actor,  
                                  save_prefix ='image', save_format ='jpg'): 
            i += 1
            if i > 5: 
                break
```
